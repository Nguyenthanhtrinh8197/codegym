```python
import turtle
import math
t= turtle.Turtle()
t.speed(5)
t.pensize(3)
t.pencolor('blue')
facesize = 200
t.penup()
t.goto(0,-200)
t.pendown()
t.circle(facesize)
#ve 2 mat
t.fillcolor('red')
t.penup()
t.goto(-100,50)
t.pendown()
eyesize = 17.5
t.begin_fill()
t.circle(eyesize)
t.end_fill()
t.penup()
t.goto(100,50)
t.pendown()
t.begin_fill()
t.circle(eyesize)
t.end_fill()
#ve nu cuoi
t.penup()
t.goto(-100,-70)
t.pendown()
t.rt(90)
t.circle(100,180)
turtle.done()
```
