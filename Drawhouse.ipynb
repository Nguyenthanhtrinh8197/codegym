```python
import turtle
import math
t = turtle.Turtle()
turtle.bgcolor('pink')
chieu_cao_nha = 60
chieu_dai_nha = 80
#vengoinha
t.fillcolor('gray')
t.begin_fill()
t.forward(chieu_cao_nha)
t.left(120)
t.forward(chieu_cao_nha)
t.left(120)
t.forward(chieu_cao_nha)
t.end_fill()
t.fillcolor('blue')
t.begin_fill()
t.right(240)
t.forward(chieu_cao_nha)
t.right(90)
t.forward(chieu_cao_nha)
t.right(90)
t.forward(chieu_cao_nha)
t.right(90)
t.forward(chieu_cao_nha)
t.right(90)
t.forward(chieu_cao_nha)
t.end_fill()
t.fillcolor('gray')
t.begin_fill()
t.left(30)
t.forward(chieu_dai_nha)
t.left(90)
t.forward(chieu_cao_nha)
t.left(90)
t.forward(chieu_dai_nha)
t.left(90)
t.forward(chieu_cao_nha)
t.end_fill()
t.fillcolor('blue')
t.begin_fill()
t.left(90)
t.forward(chieu_dai_nha)
t.right(120)
t.forward(chieu_cao_nha)
t.right(60)
t.forward(chieu_dai_nha)
t.right(120)
t.forward(chieu_cao_nha)
t.end_fill()
t.right(90)
#vematroi
khoang_cach_mat_troi = 150
t.penup()
t.forward(khoang_cach_mat_troi)
t.left(90)
t.forward(khoang_cach_mat_troi)
t.pendown()
t.fillcolor('yellow')
t.begin_fill()
t.circle(30)
t.end_fill()
#vecuachinh
do_rong_cua_chinh = 20
chieu_cao_cua_chinh = 40
Khoang_lui_theo_do_rong = khoang_cach_mat_troi + chieu_dai_nha/2
Khoang_lui_theo_do_cao = khoang_cach_mat_troi + chieu_dai_nha - chieu_cao_nha
t.penup()
t.left(90)
t.forward(Khoang_lui_theo_do_rong)
t.left(90)
t.forward(Khoang_lui_theo_do_cao)
t.pendown()
t.fillcolor('black')
t.begin_fill()
t.forward(chieu_cao_cua_chinh)
t.left(90)
t.forward(do_rong_cua_chinh)
t.left(90)
t.forward(chieu_cao_cua_chinh)
t.left(90)
t.forward(do_rong_cua_chinh)
t.end_fill()
t.right(180)
t.penup()
t.forward(do_rong_cua_chinh)
t.left(30)
t.forward(chieu_dai_nha)
t.pendown()
#vecuaso
canh_cua_so = 20
t.fillcolor('white')
t.begin_fill()
t.right(120)
t.forward(canh_cua_so)
t.right(60)
t.forward(canh_cua_so)
t.right(120)
t.forward(canh_cua_so)
t.right(60)
t.forward(canh_cua_so)
t.end_fill()
#vecaicay
chieu_rong_than_cay = 30
chieu_cao_than_cay = 80
t.penup()
t.right(30)
t.forward(200)
t.right(90)
t.forward(100)
t.pendown()
t.fillcolor('brown')
t.begin_fill()
t.right(90)
t.forward(chieu_rong_than_cay)
t.right(90)
t.forward(chieu_cao_than_cay)
t.right(90)
t.forward(chieu_rong_than_cay)
t.right(90)
t.forward(chieu_cao_than_cay)
t.end_fill()
#vechumla1
t.right(180)
t.forward(80)
t.fillcolor('green')
tan_cay = chieu_rong_than_cay*3
t.begin_fill()
t.right(90)
t.forward(chieu_rong_than_cay)
t.left(120)
t.forward(tan_cay)
t.left(120)
t.forward(tan_cay)
t.left(120)
t.forward(tan_cay)
t.end_fill()
#vechumla2
chieu_cao_tan_cay = tan_cay*math.sqrt(3)/2
t.penup()
t.left(90)
t.forward(chieu_cao_tan_cay)
t.left(90)
t.pendown()
t.fillcolor('green')
t.begin_fill()
t.right(60)
t.forward(tan_cay)
t.left(120)
t.forward(tan_cay)
t.left(120)
t.forward(tan_cay)
t.end_fill()
#vechumcay3
t.penup()
t.left(90)
t.forward(chieu_cao_tan_cay)
t.left(90)
t.pendown()
t.fillcolor('green')
t.begin_fill()
t.right(60)
t.forward(tan_cay)
t.left(120)
t.forward(tan_cay)
t.left(120)
t.forward(tan_cay)
t.end_fill()
turtle.done()
```
